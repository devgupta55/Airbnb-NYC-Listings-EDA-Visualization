**Task 4: Data Cleaning**

```python
# Total null values in each column
df.isnull().sum()

# Dropping rows with null values
df.dropna(inplace = True)
```

```python
# dealing with duplicate rows
df.duplicated().sum()

# deleting duplicate rows
df.drop_duplicates(inplace = True)
```

```python
# checking datatypes
df.dtypes
```

```python
# changing datatype
df['id'] = df['id'].astype(object)

df['host_id'] = df['host_id'].astype(object)

df.dtypes
```

#### EDA
**Task 5: Data Analysis**

**Univariate Analysis**

```python
# identifying outliers in price

df_temp = df[df['price']<1500]
plt.figure(figsize = (6,3))
sns.boxplot(data = df_temp, x = 'price')
```

```python
# price distribution 
plt.figure(figsize = (6,3))
sns.histplot(data = df_temp, x = 'price', bins = 100)
plt.title('Price Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
df.columns
```

```python
# Property Available 365 Days Distribution
plt.figure(figsize = (6,3))
plt.title('Availability 365 Days')
sns.histplot(data = df_temp, x = 'availability_365')
plt.xlabel('Availibity 365 Days')
plt.ylabel('Frequency')
plt.show()
```

```python
# Mean of Price by Neighbourhood
df_temp.groupby(by = 'neighbourhood_group')['price'].mean()
```

```python
df_temp.columns
```

**Feature Engineering**

```python
# ['price per bed']
df_temp['price per bed'] = df_temp['price']/df_temp['beds']

df.head()
```

```python
#average price per bed by neighbourhood
df_temp.groupby(by = 'neighbourhood')['price per bed'].mean()
```

**Bi Variable Analysis**
One variable depenency in another variable

```python
df_temp.columns
```

```python
# price dependency on neighbourhood
plt.figure(figsize = (6,3))
plt.title('Price by Neighbourhood and Room type')
sns.barplot(data = df_temp, x = 'price', y = 'neighbourhood_group', hue = 'room_type', errorbar = None)
plt.ylabel('Neighbourhood')
plt.xlabel('Price')
plt.show()
```

```python
# number of reviews and price relation
plt.figure(figsize = (6,3))
plt.title("Locality and Review Dependency")
sns.scatterplot(data = df_temp, x = 'number_of_reviews', y = 'price', hue = 'neighbourhood_group')
plt.xlabel('Number of Reviews')
plt.ylabel('Price ($)')
plt.show()
```

```python
# pair plot of numerical columns
sns.pairplot(data=df_temp, vars=['price', 'minimum_nights', 'number_of_reviews', 'availability_365'], hue='room_type')
plt.show()
```

```python
#Geographical Distribution of AirBnb Listing
plt.figure(figsize = (6,3))
sns.scatterplot(data=df_temp, x='longitude', y='latitude', hue='room_type')
plt.title("Geographical Distribution of AirBnb Listing")
plt.show()
```

```python
# heat map - correlation of one variable with others for numerical column

corr = df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()
corr

plt.figure(figsize=(8, 6))
sns.heatmap(data=corr, annot=True)
```
